```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example data: replace with your experimental data
# Concentration (mg/L)
concentrations = np.array([2, 4, 6, 8, 10])

# Absorbance values
absorbance = np.array([0.15, 0.30, 0.45, 0.60, 0.75])

# Reshape data for linear regression
X = concentrations.reshape(-1, 1)  # Independent variable (concentration)
y = absorbance                   # Dependent variable (absorbance)

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Get the slope (m) and intercept (b) for the equation y = mx + b
slope = model.coef_[0]
intercept = model.intercept_

# Display the calibration curve equation
print(f"Calibration Curve Equation: Absorbance = {slope:.4f} * Concentration + {intercept:.4f}")

# Predict absorbance for the given concentrations (optional)
predicted_absorbance = model.predict(X)

# Plot the calibration curve
plt.scatter(concentrations, absorbance, color='blue', label='Experimental Data')
plt.plot(concentrations, predicted_absorbance, color='red', label=f'Fit: y = {slope:.4f}x + {intercept:.4f}')
plt.title("Calibration Curve for Potassium Dichromate")
plt.xlabel("Concentration (mg/L)")
plt.ylabel("Absorbance")
plt.legend()
plt.grid()
plt.show()
```

```python
# for this make it aborbance on y axis and concentration on x axis and also make sure the image shown is saved under the folder /img create if not there also for the slightly scattered blue point join using a curved line in one image and without in the other image show botha and save to /img folder import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.optimize import curve_fit


folder = 'img/'

# Potassium Di Chromate (K2Cr207)
# Calibration Curve Data
calibration_data = {
    "Absorbance": [0, 0.15, 0.35, 0.69, 0.95, 0.98],  # Absorbance Values
    "Concentration (mg/mL)": [0 ,0.2 ,0.4 ,0.6 ,0.8,1]  # Concentration Values
}

calib_df = pd.DataFrame(calibration_data)

# Perform Linear Regression (y = mx + b)
slope, intercept, r_value, _, _ = linregress(calib_df["Absorbance"], calib_df["Concentration (mg/mL)"])

# Print Calibration Equation
print(f"Calibration Curve Equation: Concentration = ({slope:.4f} * Absorbance) + {intercept:.4f}")


# Coefficient of determination
# Shows how well the linear model fits the data.
# R=1 → Perfect positive correlation (ideal case, straight-line fit).
# R=−1 → Perfect negative correlation (not applicable here).
# R=0 → No correlation (random relationship).

print(f"R-squared: {r_value**2:.4f}")

# Plot Calibration Curve
plt.figure(figsize=(8, 5))
plt.scatter(calib_df["Absorbance"], calib_df["Concentration (mg/mL)"], color='blue', label="Calibration Data")
plt.plot(calib_df["Absorbance"], slope * calib_df["Absorbance"] + intercept, color='red', linestyle='--', label=f"Fit: y={slope:.4f}x + {intercept:.4f}")
plt.xlabel("Absorbance")
plt.ylabel("Concentration (mg/mL)")
plt.title("Calibration Curve: Absorbance vs. Concentration (Pottasium dichromate)")
plt.legend()
plt.grid()
plt.show()
```
